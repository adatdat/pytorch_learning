# Tensor

Tensor là một mảng đa chiều (được gọi là n chiều) có thể được sử dụng để biểu diễn các vector và ma trận. Tensor có thể được sử dụng để lưu trữ các giá trị số nguyên và số thực. Nó cũng có thể được sử dụng để lưu trữ các chuỗi ký tự (được gọi là tensor ký tự). Một tensor có thể được tạo ra bằng cách sử dụng hàm torch.tensor().

```python
# Ví dụ tạo tensor với 1 gia trị
# tensor khác với Tensor (Tensor là class, tensor là function)
t = torch.tensor(2) 
display(t)
# ndim: number of dimensions
t.ndim
# item: get value of tensor
t.item()
```

```python
# Tạo tensor với 1 vector (mảng 1 chiều)
vector = torch.tensor([7, 4])
display(vector)
# ndim = 1 (1 chiều)
vector.ndim
# shape: lấy shape của vector
vector.shape
```

```python
# Tạo tensor với 1 matrix (mảng 2 chiều)
matrix = torch.tensor([[1, 2, 3], 
                       [4, 5, 6]])
display(matrix)
#2 chiều
matrix.ndim
matrix.shape
```

```python
# Tạo scalar với 1 matrix (mảng 3 chiều)
# 3 chiều và shape là (1, 3, 3) (1 matrix 3x3)
# Nhận biết số chiều của tensor bằng số dấu ngoặc vuông
sale_data = torch.tensor([[[1, 2, 3],
                        [3, 6, 9],
                        [2, 4, 5]]])
display(sale_data)
# 3 chiều và shape là (1, 3, 3)
sale_data.ndim, sale_data.shape
```

Hình dạng của tensor điển hình

```python
# Tạo random tensor với size là 3x4
# random tensor là tensor có các giá trị ngẫu nhiên
# size: shape của tensor
# random: giá trị ngẫu nhiên
# randn: giá trị ngẫu nhiên theo phân phối chuẩn
# rand: giá trị ngẫu nhiên theo phân phối đều
# randint: giá trị ngẫu nhiên theo phân phối đều trong khoảng
# randperm: giá trị ngẫu nhiên theo phân phối đều trong khoảng
# normal: giá trị ngẫu nhiên theo phân phối chuẩn
# randn_like: giá trị ngẫu nhiên theo phân phối chuẩn với shape giống tensor đầu vào
# rand_like: giá trị ngẫu nhiên theo phân phối đều với shape giống tensor đầu vào
# randint_like: giá trị ngẫu nhiên theo phân phối đều trong khoảng với shape giống tensor đầu vào
# randperm_like: giá trị ngẫu nhiên theo phân phối đều trong khoảng với shape giống tensor đầu vào
# normal_like: giá trị ngẫu nhiên theo phân phối chuẩn với shape giống tensor đầu vào
# uniform: giá trị ngẫu nhiên theo phân phối đều trong khoảng
# full: tạo tensor với giá trị nhất định
# full_like: tạo tensor với giá trị nhất định với shape giống tensor đầu vào
# arange: tạo tensor với giá trị từ start đến end với step
# linspace: tạo tensor với giá trị từ start đến end với số lượng giá trị
# logspace: tạo tensor với giá trị từ start đến end với số lượng giá trị theo cấp số nhân
# ones: tạo tensor với giá trị 1
# ones_like: tạo tensor với giá trị 1 với shape giống tensor đầu vào
# zeros: tạo tensor với giá trị 0
# zeros_like: tạo tensor với giá trị 0 với shape giống tensor đầu vào
# eye: tạo tensor với giá trị 1 trên đường chéo chính
# empty: tạo tensor với giá trị rác
# empty_like: tạo tensor với giá trị rác với shape giống tensor đầu vào
# tensor: tạo tensor với giá trị nhất định
# as_tensor: tạo tensor với giá trị nhất định
# from_numpy: tạo tensor từ numpy array
# ones_like: tạo tensor với giá trị 1 với shape giống tensor đầu vào

print('Random tensor với size là 3x4====================')
random_tensor = torch.randn(size=(3, 4))
display(random_tensor, random_tensor.dtype)
```

```python
print('Random tensor với giá trị đều bằng 0====================')
zeros_tensor = torch.zeros(size=(3, 4))
display(zeros_tensor, zeros_tensor.dtype)
```

```python
print('Random tensor với giá trị đều bằng 1====================')
ones_tensor = torch.ones(size=(3, 4))
display(ones_tensor, ones_tensor.dtype)
```

```python
print('Random tensor với giá trị từ 0 đến 9====================')
zero_to_ten = torch.arange(start=0, end=10, step=1)
display(zero_to_ten, zero_to_ten.dtype)
```

```python
print('Tạo tensor với giá trị 0 với shape giống tensor đầu vào====================')
ten_zeros = torch.zeros_like(zero_to_ten)
display(ten_zeros, ten_zeros.dtype)
```

```python
# Kiểu dữ liệu phổ biến trong PyTorch là float32 và float
# Ngoài ra còn có float16, float64, int32, int64, uint8, ...
# Kiểu dữ liệu của tensor có thể thay đổi bằng cách gọi hàm .to()
# Cũng có thể thay đổi giá trị dtype của tensor bằng hàm .type()

# Giá trị mặc định của kiểu dữ liệu trong tensor là float32
float32_tensor = torch.tensor([1.0, 2.0], 
                              dtype=None, # mặc định None là float32
                              device=None, # mặc định None là CPU
                              requires_grad=False) # mặc định False, nếu là True thì các phép tính trên tensor này sẽ được lưu lại để tính đạo hàm
display(float32_tensor, float32_tensor.dtype, float32_tensor.device)
```

```python
# Tạo tensor với kiểu dữ liệu là float16
float16_tensor = torch.ones(size=(3, 4), dtype=torch.float16)
display(float16_tensor, float16_tensor.dtype)
```

```python
# Chuyển kiểu dữ liệu của tensor thành float16
print('Chuyển kiểu dữ liệu của tensor thành float16====================')
float16_tensor = float32_tensor.to(torch.float16)
display(float16_tensor, float16_tensor.dtype, float16_tensor.device)

# Chuyển kiểu dữ liệu của tensor thành float64
print('Chuyển kiểu dữ liệu của tensor thành float64====================')
float64_tensor = float32_tensor.to(torch.float64)
display(float64_tensor, float64_tensor.dtype, float64_tensor.device)

# Chuyển kiểu dữ liệu của tensor thành int32
print('Chuyển kiểu dữ liệu của tensor thành int32====================')
int32_tensor = float32_tensor.to(torch.int32)
display(int32_tensor, int32_tensor.dtype, int32_tensor.device)

# Chuyển kiểu dữ liệu của tensor thành int64
print('Chuyển kiểu dữ liệu của tensor thành int64====================')
int64_tensor = float32_tensor.to(torch.int64)
display(int64_tensor, int64_tensor.dtype, int64_tensor.device)

# Chuyển kiểu dữ liệu của tensor thành uint8
print('Chuyển kiểu dữ liệu của tensor thành uint8====================')
uint8_tensor = float32_tensor.to(torch.uint8)
display(uint8_tensor, uint8_tensor.dtype, uint8_tensor.device)

# Chuyển kiểu dữ liệu của tensor thành bool
print('Chuyển kiểu dữ liệu của tensor thành bool====================')
bool_tensor = float32_tensor.to(torch.bool)
display(bool_tensor, bool_tensor.dtype, bool_tensor.device)

# Chuyển kiểu dữ liệu của tensor thành double
print('Chuyển kiểu dữ liệu của tensor thành double====================')
double_tensor = float32_tensor.to(torch.double)
display(double_tensor, double_tensor.dtype, double_tensor.device)

# Chuyển kiểu dữ liệu của tensor thành half
print('Chuyển kiểu dữ liệu của tensor thành half====================')
half_tensor = float32_tensor.to(torch.half)
display(half_tensor, half_tensor.dtype, half_tensor.device)

# Chuyển kiểu dữ liệu của tensor thành short
print('Chuyển kiểu dữ liệu của tensor thành short====================')
short_tensor = float32_tensor.to(torch.short)
display(short_tensor, short_tensor.dtype, short_tensor.device)

# Chuyển kiểu dữ liệu của tensor thành long
print('Chuyển kiểu dữ liệu của tensor thành long====================')
long_tensor = float32_tensor.to(torch.long)
display(long_tensor, long_tensor.dtype, long_tensor.device)

# Chuyển kiểu dữ liệu của tensor thành bfloat16
print('Chuyển kiểu dữ liệu của tensor thành bfloat16====================')
bfloat16_tensor = float32_tensor.to(torch.bfloat16)
display(bfloat16_tensor, bfloat16_tensor.dtype, bfloat16_tensor.device)

# Chuyển kiểu dữ liệu của tensor thành float
print('Chuyển kiểu dữ liệu của tensor thành float====================')
float_tensor = float32_tensor.to(torch.float)
display(float_tensor, float_tensor.dtype, float_tensor.device)

# Chuyển kiểu dữ liệu của tensor thành int
print('Chuyển kiểu dữ liệu của tensor thành int====================')
int_tensor = float32_tensor.to(torch.int)
display(int_tensor, int_tensor.dtype, int_tensor.device)

```

```python
# Ví dụ nhân 2 ma trận với nhau
# Lưu y: số cột của ma trận 1 phải bằng số hàng của ma trận 2
matrix1 = torch.tensor([[1, 2, 3],
                        [4, 5, 6]])
matrix2 = torch.tensor([[7, 8],
                        [9, 10],
                        [11, 12]])
# Dấu @ là toán tử nhân ma trận trong PyTorch (có thể dùng hàm torch.matmul). @ không được khuyến khích sử dụng
display(matrix1 @ matrix2)
# Hoặc
display(torch.matmul(matrix1, matrix2))
```

torch.matmul() có tốc độ nhanh hơn

```python
# Trong trường hợp số cột của ma trận 1 không bằng số hàng của ma trận 2 thì ta có thể dùng transpose để chuyển vị ma trận
matrix1 = torch.tensor([[1, 2, 3],
                        [4, 5, 6]])
matrix2 = torch.tensor([[7, 8, 9],
                        [10, 11, 12]])
display(matrix1,matrix2, matrix2.T)
```

```python
# Thao tác hoạt động trên matrix2 sau khi chuyển vị
print(f"Dạng gốc của matrix1: {matrix1.shape}, matrix2: {matrix2.shape}")
print(f"Dạng sau khi chuyển vị của matrix2: {matrix2.T.shape}")
print(f"Kết quả của phép nhân {matrix1.shape} * {matrix2.T.shape} <- kích thước phù hợp")
display(torch.matmul(matrix1, matrix2.T))
print(f"shape của kết quả: {torch.matmul(matrix1, matrix2.T).shape}")
```

```python
# Có thể dùng hàm torch.mm để nhân 2 ma trận với nhau thay cho @ hoặc torch.matmul
display(torch.mm(matrix1, matrix2.T))
```

# Neural Networks
Neural Neworks là một tập hợp các hàm toán học được kết nối với nhau theo một cấu trúc nhất định để giải quyết các bài toán liên quan đến Machine Learning và Deep Learning như phân loại, nhận diện, dự đoán, ... 
Một Neural Network bao gồm các thành phần chính sau:
- Input: là dữ liệu đầu vào của mô hình
- Hidden layers: là các lớp ẩn của mô hình
- Output: là dữ liệu đầu ra của mô hình

```python

# torch.nn.Linear() là một lớp trong PyTorch giúp ta tạo ra một neural network với các trọng số và hệ số điều chỉnh được khởi tạo ngẫu nhiên. Được biết như là feedforward layer, fully connected layer, hoặc affine transformation. Thực hiện phép nhân ma trận giữa input và weight, sau đó cộng với bias với công thức sau:
# y = xA^T + b 
# Trong đó:
# x: là input
# A: là weight matrix được tạo ngẫu nhiên, và được điều chỉnh trong quá trình học
# b: là bias được sử dụng bù đắp cho A và x
# y: là output
# Linear() có công thức giống với y = mx + b trong đại số tuyến tính

# manual_seed là hàm giúp ta tạo ra một số ngẫu nhiên nhưng cố định, giúp ta có thể tái tạo lại kết quả của mô hình
torch.manual_seed(42)
print('Tạo linear layer với 2 input và 6 output====================')
linear = torch.nn.Linear(in_features=2, # số chiều của input
                         out_features=6, # số chiều của output
                            bias=True) # có sử dụng bias hay không
tensor_input = torch.tensor([[1.0, 2.0],
                            [3.0, 4.0]]) # input có shape là (2, 2)
output = linear(tensor_input) # output có shape là (2, 6)
print(f"Output của linear layer: {output}\n và shape là {output.shape}")
```

#### Phép nhân ma trận rất quan trọng trong deep learning, machine learning và toán học. 

```python
# Một số phép toán tổng hợp trong tensor: torch.mean(), torch.max(), ...
x = torch.arange(0, 100, 10)
display(x)

print(f"Minimum: {x.min()}")
print(f"Maximum: {x.max()}")
# print(f"Mean: {x.mean()}") # mean() không hoạt động với kiểu dữ liệu int 
print(f"Mean: {x.type(torch.float32).mean()}") # mean() hoạt động với kiểu dữ liệu float
print(f"Sum: {x.sum()}")

# Tương tự 
torch.max(x), torch.min(x), torch.mean(x.type(torch.float32)), torch.sum(x)
```

```python
# Vị trí của giá trị lớn nhất và giá trị nhỏ nhất trong tensor
# argmax() và argmin() chỉ hoạt động với tensor có 1 chiều
# Nếu tensor có nhiều chiều thì ta phải dùng hàm torch.argmax() và torch.argmin()
# Ví dụ
# Tạo tensor với 1 chiều
tensor = torch.arange(10, 100, 10)
print(f"Tensor: {tensor}")

# Tìm vị trí của giá trị lớn nhất và giá trị nhỏ nhất trong tensor
print(f"Vị trí của giá trị lớn nhất: {tensor.argmax()}")
print(f"Vị trí của giá trị nhỏ nhất: {tensor.argmin()}")
```

```python
# Reshaping, stacking, squeezing and unsqueezing tensors trong PyTorch là những phép biến đổi tensor thành các tensor khác với shape khác nhau
# Ví dụ
# Tạo tensor với 1 chiều
tensor = torch.arange(10, 100, 10)
print(f"Tensor: {tensor} và shape là {tensor.shape}")

# reshaping là phép biến đổi tensor thành tensor khác với shape khác nhau
x_reshaped = tensor.reshape(1, 9)
print(f"Reshaped tensor: {x_reshaped} và shape là {x_reshaped.shape}")
```

```python
# view là một hàm khác để reshape tensor
x_reshaped = tensor.view(1, 9)
print(f"Reshaped tensor: {x_reshaped} và shape là {x_reshaped.shape}")
```

```python
# Stack là hàm dùng để nối các tensor với nhau
x_stacked = torch.stack([tensor, tensor, tensor], dim=0) # dim=0 là stack theo chiều dọc
print(f"Stacked tensor: {x_stacked}\n và shape là {x_stacked.shape}")
```

```python
# Squeeze là hàm dùng để giảm số chiều của tensor
print(f"x_resahped: {x_reshaped}\n và shape là {x_reshaped.shape}")

# Giảm số chiều của tensor x_stacked từ 2 chiều xuống 1 chiều
x_squeezed = x_reshaped.squeeze()
print(f"Squeezed tensor: {x_squeezed}\n và shape là {x_squeezed.shape}")
```

```python
# Unsqueeze là hàm dùng đảo ngược của squeeze, dùng để tăng số chiều của tensor
print(f"x_squeezed: {x_squeezed}\n và shape là {x_squeezed.shape}")

# Tăng số chiều của tensor x_squeezed từ 1 chiều lên 2 chiều
x_unsqueezed = x_squeezed.unsqueeze(dim=0)
print(f"Unsqueezed tensor: {x_unsqueezed}\n và shape là {x_unsqueezed.shape}")
```

```python
# Permute là hàm dùng để hoán đổi vị trí các chiều của tensor
# Ví dụ
# Tạo tensor với 3 chiều
x_original = torch.randn(size=(3, 4, 5))
print(f"Shape gốc là {x_original.shape}")

# Hoán đổi vị trí các chiều của tensor
x_permuted = x_original.permute(1, 2, 0)
print(f"Shape sau khi hoán vị là {x_permuted.shape}")
```

```python
# index select là hàm dùng để lấy các phần tử của tensor theo chỉ số
# Ví dụ
# Tạo tensor với 2 chiều
x = torch.tensor([[1, 2, 3],
                    [4, 5, 6]])
print(f"Tensor gốc: {x}")

# Lấy các phần tử của tensor theo chỉ số
# Lấy phần tử ở hàng 0
print(f"Phần tử ở hàng 0: {x[0]}")
# Lấy phần tử ở hàng 0, cột 1
print(f"Phần tử ở hàng 0, cột 1: {x[0, 1]}")
```

```python
# numpy là thư viện hỗ trợ tính toán khoa học trong Python và là thư viện cốt lõi của nhiều thư viện khác như Pandas, Scikit-learn, ...
# torch.from_numpy() là hàm dùng để chuyển đổi numpy array thành tensor
# Ví dụ
import numpy as np

array = np.arange(1.0, 8.0)
tensor = torch.from_numpy(array)
display(tensor, array)

# torch.Tensor.numpy() là hàm dùng để chuyển đổi tensor thành numpy array
# Ví dụ
tensor = torch.arange(1.0, 8.0)
array = tensor.numpy()
display(tensor, array)
```

```python
# Neural network bắt đầu với các giá trị ngẫu nhiên và sau đó cố gắng điều chỉnh các giá trị này bằng các thao tác với tensor như phép cộng, phép trừ, phép nhân, phép chia, ... để tìm ra các giá trị tối ưu nhất cho mô hình
# start with random numbers -> tensor operations -> try to make better (again and again and again) -> end with good numbers
# reproducibility là khả năng tái tạo lại của neural network với khi random seed là cố định
# Ví dụ
import torch
import random

# Khởi tạo random seed
RANDOM_SEED=42 # thử thay đổi giá trị của RANDOM_SEED và chạy lại code
torch.manual_seed(seed=RANDOM_SEED) 
random_tensor_C = torch.rand(3, 4)

# Phải reset lại seed mỗi lần gọi hàm rand() để có thể tái tạo lại kết quả
# Nếu không có dòng này thì các giá trị của random_tensor_D sẽ khác với random_tensor_C 
torch.random.manual_seed(seed=RANDOM_SEED) # thử comment dòng này và chạy lại code
random_tensor_D = torch.rand(3, 4)

print(f"Tensor C:\n{random_tensor_C}\n")
print(f"Tensor D:\n{random_tensor_D}\n")
print(f"Tensor C có giá trị bằng với Tensor D?:")
random_tensor_C == random_tensor_D
```

# GPU

```python
# GPU là bộ xử lý đồ họa, được sử dụng để tăng tốc độ tính toán của một số mô hình machine learning
# Thường được sử dụng để tính toán các mô hình deep learning, vì các mô hình này có số lượng tham số lớn và cần nhiều thời gian để tính toán nên việc sử dụng GPU sẽ giúp tăng tốc độ tính toán lên rất nhiều lần so với CPU thông thường (thường là 10 lần) 
# Để sử dụng GPU trong PyTorch, ta cần kiểm tra xem máy tính có GPU hay không bằng hàm torch.cuda.is_available()
# Nếu torch.cuda.is_available() trả về False thì ta cần cài đặt CUDA trên máy tính
import torch
torch.cuda.is_available()
```

```python
# Sau khi kiểm tra có GPU hay không, ta cần tạo một biến để lưu trữ thông tin về GPU bằng hàm torch.device() và truyền vào tên của GPU (có thể lấy thông tin về GPU bằng hàm torch.cuda.get_device_name()) 
# Ví dụ 
# Tạo biến để lưu trữ thông tin về GPU
device = "cuda" if torch.cuda.is_available() else "cpu"
device
# Nếu device là cuda thì ta có thể sử dụng GPU để tính toán, nếu là cpu thì ta sẽ sử dụng CPU để tính toán
```

```python
# Đếm số GPU có trên máy tính bằng hàm torch.cuda.device_count()
torch.cuda.device_count()
```

```python
# Đặt tensor lên GPU bằng hàm .to()
# Ví dụ
# Tạo tensor với 1 chiều
tensor = torch.tensor([1, 2, 3])

# Đặt tensor lên GPU
# cuda:0 là tên của GPU (nếu có nhiều GPU thì có thể có cuda:1, cuda:2, ...) 
tensor_on_gpu = tensor.to(device)
tensor_on_gpu
```

```python
# Đặt tensor lên CPU bằng numpy() hoặc cpu()

# Chuyển tensor thành numpy array
# Nếu tensor đang ở GPU thì ta cần chuyển về CPU trước khi chuyển thành numpy array bằng hàm .cpu().numpy() 
tensor_on_gpu.numpy()
```
